```python
data_native = pd.read_csv('/Users/catherinebreen/Documents/Chapter 1/WRRsubmission/data/snowlabels_nativeRes_labels_clean.csv')
all_HT_files = glob.glob('/Users/catherinebreen/Documents/Chapter 1/WRRsubmission/data/native_res_results/HT/*.csv')
all_CNNkeypoint_files = glob.glob('/Users/catherinebreen/Documents/Chapter 1/WRRsubmission/data/native_res_results/CNNkeypoint/predictions/*.csv')
all_CNNht_files = glob.glob('/Users/catherinebreen/Documents/Chapter 1/WRRsubmission/data/native_res_results/CNNht/**/*.csv')
```

```python
print(len(all_HT_files))
print(len(all_CNNkeypoint_files))
print(len(all_CNNht_files))
```

```python
## concatenate all the rows within all the dataframes
HT_data = pd.concat([pd.read_csv(f) for f in all_HT_files ], ignore_index=True)
HT_data['filename'] = HT_data['Var1']
CNNkeypoint_data = pd.concat([pd.read_csv(f) for f in all_CNNkeypoint_files ], ignore_index=True)
CNNht_data = pd.concat([pd.read_csv(f) for f in all_CNNht_files ], ignore_index=True)
```

```python
data_native = data_native.merge(HT_data, how="left", on="filename")
data_native = data_native.merge(CNNkeypoint_data, how="left", on="filename")
data_native = data_native.merge(CNNht_data, how="left", on="filename")
```

```python
### update columns
data_native['HT_px'] = data_native['Var2']
data_native['keypoint_px'] = data_native['total_length_pixel_x']
data_native['CNNht_px'] = data_native['total_length_pixel_y']
```
