```python

import matplotlib.pyplot as plt
import numpy as np

x = ['January', ' February', 'March', 'April']
y1 = [-6.8, -6.8, -3.3, 0.8]
y2 = [-1.8, -0.9, 3.5, 9.1]
```

```python
fig, ax = plt.subplots()

twin1 = ax.twinx()
p2, = twin1.plot(x, [2.0, 1.9, 1.4,0.4], "r-", label="snow depth (m)")
twin1.set_ylim(-3,3)
twin1.set_ylabel('snow depth (m)')
twin1.yaxis.label.set_color(p2.get_color())

ax.fill_between(x, y1, y2)
ax.set_ylim(-10, 10)
ax.axhline(y=0.0, color='black', linestyle='-')
ax.set_ylabel('temperature (Celsius)')

for i, txt in enumerate(y1):
    ax.annotate(y1[i], (x[i], y1[i]))
    ax.annotate(y2[i], (x[i], y2[i]))
```
